```python
class Kuyruk :
    kuyruk = [] # FIFO 
    def kuyruga_ekle( self,x : int ) : 
        if self.kuyruk.count(x)>0:
            return
        self.kuyruk.append(x)
    def kuyruk_uzunluk(self): 
        return len(self.kuyruk)
    def kuyrugu_isle(self): 
        if kuyruk_uzunluk() == 0:
            print("kuyruk boş")
            return
        print(self.kuyruk , self.kuyruk[0])
        self.kuyruk.pop(0)
```

```python
class Yigin : # stack
    yigin = [] # LIFO
    def yigina_ekle(self,x:int):
        self.yigin.append(x)
    def yigini_isle(self):
        if len(self.yigin) == 0:
            print("yığın boş")
            return
        print(self.yigin, self.yigin[-1])
        self.yigin.pop()
```

```python
g = {
    1:[2,3],
    2:[5],
    3:[4],
    5:[4]
}
y = Yigin()
```
