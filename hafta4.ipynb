```python
import heapq 
 
def dijkstra(graph, start): 
    # Initialize distances and priority queue 
    distances = {node: float('inf') for node in graph} 
    distances[start] = 0 
    queue = [(0, start)] 
 
    while queue: 
        current_distance, current_node = heapq.heappop(queue) 
 
        # Skip if we already found a shorter path 
        if current_distance > distances[current_node]: 
            continue 
 
        for neighbor, weight in graph[current_node].items(): 
            distance = current_distance + weight 
            if distance < distances[neighbor]: 
                distances[neighbor] = distance 
                heapq.heappush(queue, (distance, neighbor)) 
 
    return distances 
 
# Example graph 
graph = { 
    'A': {'B': 1, 'C': 4}, 
    'B': {'A': 1, 'C': 2, 'D': 5}, 
    'C': {'A': 4, 'B': 2, 'D': 1}, 
    'D': {'B': 5, 'C': 1} 
} 
 
print(dijkstra(graph, 'A')) 
```

Bubble Sort algorithm includes the following steps:

1. Set N = Length of Array
2. Array[j] > Array[j + 1] (Decision)
    if no - continue with the step 4
3. Swap Array[j] and Array[j+1]
4. Increment j (j++)
5. j < N - i - 1 (Decision)
    if yes - continue with the step 2
6. Clear j (j = 0)
7. Increment i (i++)
8. i < N - 1 (Decision)
    if yes continue with the step 2
9. Ordered Array 

```python
dizi = [6,9,3,5,2,7,1]

N = len(dizi)
i=0
j=0

while i < N -1 :
    while j < N - i - 1:
        if dizi[j] > dizi[j+1] :
            # takas yap
            tmp = dizi[j]
            dizi[j] = dizi[j+1]
            dizi[j+1] = tmp
        j += 1
    j=0
    i+=1

dizi
```

```python
def faktoriyel(x:int):
    if x < 0 :
        print("0 dan küçük olamaz.")
        return
    f = x
    for i in range(1,x):
        f = f*i
    return f

faktoriyel(5)
```

```python
def faktoriyel_yinelemeli(x:int):
    if x < 0 :
        print("hatalı")
        return
    if x == 0 or x == 1 :
        return 1
    else:
        return x * faktoriyel_yinelemeli(x-1)

faktoriyel_yinelemeli(5)
```

```python
# doğrusal arama algoritması
dizi = [6,9,3,5,2,7,1]
aranan = 5

def dogrusal_arama(dizi:list, aranan:int):
    for i in range(len(dizi)) :
        if dizi[i] == aranan:
            return i
    return -1

dogrusal_arama(dizi, aranan)
```
