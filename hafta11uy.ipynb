```python
transactions = [
    ["bread", "milk", "butter"],
    ["bread", "diapers", "beer", "eggs"],
    ["milk", "diapers", "beer", "cola"],
    ["bread", "milk", "diapers", "beer"],
    ["bread", "milk", "diapers", "cola"],
    ["milk", "bread", "butter", "jam"],
    ["bread", "butter"],
    ["milk", "diapers", "beer", "bread"],
]
```

```python
# her üründen kaç tane olduğunu sayalım
essiz_urunler = [] # unique - tekil - biricik - eşsiz
essiz_urun_sayisi = []
for fis in transactions:
    for urun in fis:
        urun_var = False
        for u in essiz_urunler:
            if u == urun :
                urun_var = True
                continue
        if not urun_var :
            essiz_urunler.append(urun)
            essiz_urun_sayisi.append(1)
        else:
            indis = essiz_urunler.index(urun)
            essiz_urun_sayisi[indis] += 1

essiz_urunler , essiz_urun_sayisi
```
