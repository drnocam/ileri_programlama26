```python
# siyah beyaz resim ve edge detection vb. transpose

x = [165,170,172,174,175,176,178,180,182,188]

x
```

```python
x_ort = sum(x)/len(x) 

x_var = sum([(i-x_ort)**2 for i in x])/(len(x)-1)

x_stdev = x_var**0.5

print(f"x ortalama: {x_ort:.2f}, x varyans: {x_var:.2f}, x standart sapma: {x_stdev:.2f}")
```

```python
# boyu 185cm olan bir ki艧inin z-skoru nedir?
z_score = (195 - x_ort) / x_stdev
z_score
```

```python
from scipy import stats
import numpy as np

# Ortalama
scipy_ort = np.mean(x)

# Varyans (sample)
scipy_var = stats.tvar(x)

# Standart sapma (sample)
scipy_stdev = stats.tstd(x)

# Z-score (195 cm i莽in)
scipy_z_score = (195 - scipy_ort) / scipy_stdev

# T眉m x de臒erleri i莽in z-scores
all_z_scores = stats.zscore(x, ddof=1)

# Z-score'dan y眉zde oran谋 (percentile) hesaplama
scipy_percentile = stats.norm.cdf(scipy_z_score) * 100
all_percentiles = [stats.norm.cdf(z) * 100 for z in all_z_scores]

print(f"Ortalama: {scipy_ort:.2f}")
print(f"Varyans: {scipy_var:.2f}")
print(f"Standart Sapma: {scipy_stdev:.2f}")
print(f"195 cm i莽in Z-Score: {scipy_z_score:.2f}")
print(f"195 cm i莽in Y眉zdelik Dilim (Percentile): {scipy_percentile:.2f}%")
print(f"\nT眉m x de臒erleri i莽in Z-Scores:")
for val, z in zip(x, all_z_scores):
    print(f"  {val} cm -> z-score: {z:.4f}")
```

```python
import numpy as np
import pandas as pd
f = open("titanic.csv","r")

df = pd.read_csv(f)
np_age = df['Age'].dropna().to_numpy(dtype=float)
# print("\nPandas'dan numpy dizisine Age aktar谋m谋:")
# print(np_age)
# print("Mean age (numpy):", np_age.mean())
# print("Max age (numpy):", np_age.max())
# print("Min age (numpy):", np_age.min())
age_var = np.var(np_age, ddof=1)
age_stdev = np.sqrt(age_var)
print(f"Age varyans: {age_var:.2f}, Age standart sapma: {age_stdev:.2f}")
```

```python
metrekareler = [80, 100, 120,135, 150, 165,180,200,220,250]
fiyatlar = [1200,1500,1850,2100,2300,2600,2850,3200,3500,4000]
```

```python
ort_metrekare = np.mean(metrekareler)
ort_fiyat = np.mean(fiyatlar)

print(f"Ortalama metrekare: {ort_metrekare:.2f}, Ortalama fiyat: {ort_fiyat:.2f}")

corelation = np.corrcoef(metrekareler, fiyatlar)[0, 1]
print(f"Metrekareler ve fiyatlar aras谋ndaki korelasyon: {corelation:.5f}")
```

```python
# y = ax + b i莽in a ve b hesaplama

a = np.sum((np.array(metrekareler) - ort_metrekare) * (np.array(fiyatlar) - ort_fiyat)) / np.sum((np.array(metrekareler) - ort_metrekare) ** 2)
b = ort_fiyat - a * ort_metrekare

print(f"a (e臒im): {a:.4f}")
print(f"b (sabit terim): {b:.4f}")
print(f"Model: fiyat = {a:.4f} * metrekare + {b:.4f}")
```

```python
metrekare = 230
model_fiyat_hesab谋 = metrekare * a + b
print(f"{metrekare} metrekare i莽in model taraf谋ndan tahmin edilen fiyat: {model_fiyat_hesab谋:.2f}")
```
