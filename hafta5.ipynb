```python

kurallar = {
    "ömer":"insan",
    "insan":"canlı"
}

def is_canli(cat):
    global kurallar
    c = kurallar[cat] 
    if c=="canlı":
        return True
    d = kurallar[c]
    if d == "canlı":
        return True
    return False
        

print( is_canli("ömer") )
```

```python
satislar = [
    [3,7,2],
    [5,3,1],
    [2,7,1]
] # 3 satır 3 sütun
fiyatlar =[
    [50],
    [70],
    [150]
] # 3 satır 1 sütun

# satislar x fiyatlar = toplam gelir
# 3x50 + 7x70 + 2x150
# satislar[0][0] x fiyatlar[0][0] + satislar[0][1] x fiyatlar[1][0] + satislar[0][2] x fiyatlar[2][0]
# 5*50 + 3*70 + 1*150
# satislar[1][0]*fiyatlar[0][0] + satislar[1][1]*fiyatlar[1][0] + satislar[1][2]*fiyatlar[2][0]
toplam = 0
for i in range(len(satislar)):
    for j in range(len(satislar[i])):
        for t in range(len(fiyatlar[0])):
            toplam += satislar[i][j]*fiyatlar[j][t]
toplam 
```
