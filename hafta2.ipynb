```python
class AgacVeriTipi:
    __mesafe = 0
    def __init__(self, etiket=""):
        self.__etiket = etiket
        print("sınıf oluşturuldu.")
    def __del__(self):
        print("sınıf silindi.")
    def get_etiket(self):
        return self.__etiket
    def set_etiket(self, value):
        self.__etiket = value
    etiket = property(get_etiket, set_etiket)

# Yeni nesne oluşturma ve test
agacyapisi = AgacVeriTipi("A")
print(agacyapisi.get_etiket())
# print(agacyapisi.__etiket)
agacyapisi.__mesafe = 10
print(agacyapisi.__mesafe)
agacyapisi.set_etiket("B")
print(agacyapisi.etiket)
```

```python
class TreeNode:
    """İkili ağaç düğümü"""
    def __init__(self, deger):
        self.deger = deger
        self.sol = None
        self.sag = None
    
    def __str__(self):
        return f"TreeNode({self.deger})"
    
    def __repr__(self):
        return self.__str__()

class BinaryTree:
    """İkili ağaç sınıfı"""
    def __init__(self, kok_deger=None):
        self.kok = TreeNode(kok_deger) if kok_deger else None
    
    def ekle(self, deger):
        """Ağaca yeni düğüm ekle (BST kuralına göre)"""
        if self.kok is None:
            self.kok = TreeNode(deger)
        else:
            self._ekle_recursive(self.kok, deger)
    
    def _ekle_recursive(self, dugum, deger):
        if deger < dugum.deger:
            if dugum.sol is None:
                dugum.sol = TreeNode(deger)
            else:
                self._ekle_recursive(dugum.sol, deger)
        else:
            if dugum.sag is None:
                dugum.sag = TreeNode(deger)
            else:
                self._ekle_recursive(dugum.sag, deger)
        
    def __str__(self):
        return f"BinaryTree(kök={self.kok})"

# Kullanım
bt = BinaryTree()
for deger in [5, 3, 7, 2, 4, 6, 8]:
    bt.ekle(deger)
    def print_tree_text(node, indent="", last=True):
        if node is None:
            print(indent + ("└── " if last else "├── ") + "None")
            return
        print(indent + ("└── " if last else "├── ") + str(node.deger))
        indent += "    " if last else "│   "
        if node.sag or node.sol:
            if node.sag:
                print_tree_text(node.sag, indent, False)
            else:
                print(indent + "├── None")
            if node.sol:
                print_tree_text(node.sol, indent, True)
            else:
                print(indent + "└── None")

    print_tree_text(bt.kok)
print(bt)
```
